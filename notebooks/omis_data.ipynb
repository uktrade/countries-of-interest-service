```python
def peek_at_table(connection, table, n_rows=10, schema='public'):
    sql = '''
    select * from {schema}."{table}" limit {n_rows}
    '''.format(schema=schema, table=table, n_rows=n_rows)
    cursor = connection.cursor()
    cursor.execute(sql)
    rows = cursor.fetchall()
    columns = [d[0] for d in cursor.description]
    df = pd.DataFrame(rows, columns=columns)
    return df
```

#### omis tables

```python
sql = ''' 
select 
  table_name 
  
from information_schema.tables 

where table_schema='public' and 
  table_name ilike 'omis%' 
  
order by table_name 

'''
cursor.execute(sql)
tables = [r[0] for r in cursor.fetchall()]
tables
```

```python
for t in tables:
    print('table: {}'.format(t))
    df = peek_at_table(connection, t)
    df = df[sorted(df.columns)]
    display(df.head())
```

### omis tables with a country field

```python
for t in tables:
    print('table: {}'.format(t))
    df = peek_at_table(connection, t)
    df = df[sorted(df.columns)]
    df = df[[c for c in df.columns if 'country' in c]]
    display(df.head())
```
